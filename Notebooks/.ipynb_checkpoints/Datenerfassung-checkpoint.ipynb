```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import random

%matplotlib inline

# Pfad der Daten
DIRECTORY = "F:/Machine Learning Data/PKLot/PKLotSegmented"

# Ordner der Daten; der Index entspricht dem Label
PARKING_LOT = ["Empty", "Occupied"]

# Festlegen der Größe der Images
IMG_SIZE_X = 50
IMG_SIZE_Y = 50

training_data = []



# Trainingsdaten erzeugen
for parkinglot in PARKING_LOT: # Durchgehen der Pfade
    path = os.path.join(DIRECTORY, parkinglot) # Aktueller Pfad
    label = PARKING_LOT.index(parkinglot) # Aktuelles label
    for img in os.listdir(path): # Jedes Image im aktuellen Pfad durchgehen
        img = cv2.imread(os.path.join(path, img)) # Image als Array einlesen
        scaled = cv2.resize(img, (IMG_SIZE_X, IMG_SIZE_Y)) # Skalieren des Images
        training_data.append([scaled, label]) # Image-Array zu Trainingsdaten hinzufügen 
        plt.imshow(scaled)
        plt.show()
        break
    break
```

```python
# Zum optimalen Lernen werden die Daten gemischt

random.shuffle(training_data)
```

```python
train_data = []
train_label = []

for features, label in training_data:
    train_data.append(features)
    train_label.append(label)

    
# Trainingsdaten müssen in ein Numpy-Array umgewandelt werden
train_data = np.array(train_data).reshape(1, IMG_SIZE_X, IMG_SIZE_Y, 3)
plt.imshow(train_data[0])
plt.show()
```

```python
# Speichern der Daten
np.save('rgbfeatures.npy', train_data)
np.save('rgblabels.npy', train_label)
```
